```
#importando bibliotecas
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```
#realizando carga de dados
path = '/content/anuario-2019-abertos-tabela1.2-produção.csv'

base = pd.read_csv(path,
                   sep=';')

base.head()
```

```
base['ANO'] = pd.to_datetime(base['ANO'], format='%Y')
```

```
base['year'] = base['ANO'].dt.year

base.head()
```

```
base.info()
```

```
base['PRODUÇÃO EM mil barris / dia'] = base['PRODUÇÃO EM mil barris / dia'].str.replace(',','.')
```

```
base['PRODUÇÃO EM mil barris / dia'] = base['PRODUÇÃO EM mil barris / dia'].astype('float64')
```

```
#produção de petróleo mundial por continente entre os anos de 2010 e 2019
base['PRODUÇÃO EM mil barris / dia'].groupby(base['REGIÃO ']).mean().plot.bar();
```

```
#produção de petróleo mundial por ano
base['PRODUÇÃO EM mil barris / dia'].groupby(base['year']).mean().plot.bar();
```

```
#produção de petróleo por pais entre 2010 e 2019
sns.set(rc={'figure.figsize':(12, 6)})

base['PRODUÇÃO EM mil barris / dia'].groupby(base['PAÍS']).mean().plot.bar();
```

```
#para ver todas as linhas da coluna país
pd.set_option('display.max_rows', None)

base['PAÍS']
```

```
#produção de petróleo no Brasil por ano
plot_prod_pais = base.loc[(base['PAÍS'] == 'Brasil') | (base['PAÍS'] == 'Brasil (1)')]

plot_prod_pais['PRODUÇÃO EM mil barris / dia'].groupby(base['year']).mean().plot.bar();
```

```
#produção de petróleo nas Américas Central e do Sul entre 2010 e 2019
plot_prod_pais = base.loc[base['REGIÃO '] == 'Américas Central e do Sul']

plot_prod_pais['PRODUÇÃO EM mil barris / dia'].groupby(base['PAÍS']).mean().plot.bar();
```

```
#produção de petróleo na Venezuela por ano
plot_prod_pais = base.loc[base['PAÍS'] == 'Venezuela']

plot_prod_pais['PRODUÇÃO EM mil barris / dia'].groupby(base['year']).mean().plot.bar();
```

```
#produção de petróleo na Europa entre 2010 e 2019
plot_prod_pais = base.loc[base['REGIÃO '] == 'Europa']

plot_prod_pais['PRODUÇÃO EM mil barris / dia'].groupby(base['PAÍS']).mean().plot.bar();
```
